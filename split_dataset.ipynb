```python
import os
import pandas as pd
import shutil
```

```python
def path_check(p):
    if os.path.exists(p):
        shutil.rmtree(p)
    os.makedirs(p)
```

```python
dataset_path = './dataset/crop_o_images/'
train_excel_path = './dataset/train-glaucoma-uod-relabel.csv'
valid_excel_path = './dataset/valid-glaucoma-uod-relabel.csv'
test_excel_path = './dataset/test-glaucoma-uod-relabel.csv'
```

```python
new_dataset_path = './dataset/database-3cls/'
path_check(new_dataset_path)
```

```python
def image_move(dataset_path, excel_path, new_dataset_path, mode):
    df = pd.read_csv(excel_path)
    ds_im_path_neg = os.path.join(new_dataset_path, mode, '0_neg')
    ds_im_path_pos = os.path.join(new_dataset_path, mode, '1_pos')
    ds_im_path_sus = os.path.join(new_dataset_path, mode, '2_sus')
    path_check(ds_im_path_neg)
    path_check(ds_im_path_pos)
    path_check(ds_im_path_sus)
    
    for i in range(len(df)):
        print(df.iloc[i]['x'], int(df.iloc[i]['y']))
        sr_im_path = os.path.join(dataset_path, df.iloc[i]['x'])
        if os.path.exists(sr_im_path):
            if int(df.iloc[i]['y_']) == 0:
                shutil.copy(sr_im_path, ds_im_path_neg)
            elif int(df.iloc[i]['y_']) == 1:
                shutil.copy(sr_im_path, ds_im_path_pos)
```

```python
image_move(dataset_path, valid_excel_path, new_dataset_path, 'val')
```
